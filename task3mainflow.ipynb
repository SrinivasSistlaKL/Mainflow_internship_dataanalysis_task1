```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate a larger sample dataset
np.random.seed(0)
categories = [f'Category {i}' for i in range(1, 21)]
values1 = np.random.randint(10, 100, size=20)
values2 = np.random.randint(50, 150, size=20)

df = pd.DataFrame({
    'Category': categories,
    'Values1': values1,
    'Values2': values2
})

# Create a bar chart
plt.figure(figsize=(14, 7))
plt.bar(df['Category'], df['Values1'], color='b', label='Values1')
plt.xlabel('Category')
plt.ylabel('Values1')
plt.title('Bar Chart')
plt.xticks(rotation=45)  # Rotate category labels for better readability
plt.legend()
plt.tight_layout()
plt.show()

# Create a line chart
plt.figure(figsize=(14, 7))
plt.plot(df['Category'], df['Values2'], marker='o', linestyle='-', color='r', label='Values2')
plt.xlabel('Category')
plt.ylabel('Values2')
plt.title('Line Chart')
plt.xticks(rotation=45)  # Rotate category labels for better readability
plt.legend()
plt.tight_layout()
plt.show()
```
