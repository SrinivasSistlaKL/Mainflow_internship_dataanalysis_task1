```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data
data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June'],
    'Sales': [200, 250, 300, 150, 400, 350],
    'Expenses': [180, 200, 220, 140, 320, 300]
}

# Creating DataFrame
df = pd.DataFrame(data)

# Creating the figure and axis objects
fig, ax1 = plt.subplots(figsize=(10, 5))

# Bar chart for Sales
ax1.bar(df['Month'], df['Sales'], color='skyblue', label='Sales')

# Line chart for Expenses
ax2 = ax1.twinx()
ax2.plot(df['Month'], df['Expenses'], color='red', marker='o', label='Expenses')

# Adding labels and title
ax1.set_xlabel('Month')
ax1.set_ylabel('Sales', color='skyblue')
ax2.set_ylabel('Expenses', color='red')
plt.title('Monthly Sales and Expenses')

# Adding legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Displaying the chart
plt.show()
```
